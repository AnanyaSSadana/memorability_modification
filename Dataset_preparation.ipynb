```python
df = pd.read_csv('../Dataset/CelebA/aggan_test_classes.csv')
labels = df.sort_values('Class')
class_names = list(labels.Class.unique())
```

```python
class_names
```

```python
for i in class_names:
    os.makedirs(os.path.join('../Dataset/CelebA/dataset_aggan/test',i))
```

```python
for i in class_names:
  for c in list(df[df['Class']==i]['image_id']):    # c  ->  individual image

    if not os.path.exists('../Dataset/CelebA/dataset_aggan/test/'+i+'/'+c):
      # Creating path to the image
      get_image = os.path.join('../Dataset/CelebA/img_align_celeba/img_align_celeba_test/', c)
      # move the image to this path
      move_image = shutil.copy(get_image,'../Dataset/CelebA/dataset_aggan/test/'+i)
```
